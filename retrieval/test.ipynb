```python
from langchain_community.retrievers import BM25Retriever
import jieba
from typing import List
from nltk.corpus import stopwords
from langchain_core.documents import Document


def chinese_tokenizer(text: str) -> List[str]:
    tokens = jieba.lcut(text)
    return [token for token in tokens if token not in stopwords.words('chinese')]


def bm25retriever():
    file_path = "D:\\desktop\\1\\study\\pg\\5020\\project\\data_clean\\textbooks\\zh_paragraph\\all_books.txt"
    with open(file_path, "r", encoding="utf-8") as f:
        Docs = ([Document(page_content=i) for i in f.readlines() if i.strip() and len(i.strip()) > 10])

    bm25retriever = BM25Retriever.from_documents(
        Docs,
        k = 5,
        preprocess_func=chinese_tokenizer
    )
    return Docs, bm25retriever
```

```python
docs, retriever = bm25retriever()
```

```python
import pickle

with open("D:\\desktop\\1\\study\\pg\\5020\\project\\bm25_db\\bm25_retriever.pkl", "wb") as f:
    pickle.dump(retriever, f)
```
